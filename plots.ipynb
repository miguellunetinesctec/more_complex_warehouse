```python
# Import necessary libraries
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Set Seaborn style for better aesthetics
sns.set(style="whitegrid")

# Load the CSV file
df = pd.read_csv("results/tables/table_results.csv")

# Display the dataframe
print("=== Raw Data ===")
display(df)

# Plot: MeanReward comparison
plt.figure(figsize=(8, 5))
sns.barplot(x="Policy", y="MeanReward", data=df, palette="viridis")
plt.title("Mean Reward per Policy")
plt.ylabel("Mean Reward")
plt.xlabel("Policy")
plt.xticks(rotation=45)
plt.grid(False)
plt.tight_layout()
plt.show()

# Plot: Average DoS across Sections for each Policy
plt.figure(figsize=(10, 6))
dos_sections = df[["AvgDosSectionA", "AvgDosSectionB", "AvgDosSectionC"]]
dos_sections.index = df["Policy"]
dos_sections.plot(kind="bar", figsize=(10, 6), colormap="Set2")
plt.title("Average DoS per Section by Policy")
plt.ylabel("Average DoS")
plt.xlabel("Policy")
plt.xticks(rotation=0)
plt.legend(title="Section")
plt.grid(False)
plt.tight_layout()
plt.show()
```
